```python

#print right angled triangle
n=eval(input())

print("Right angled Pattern")
for i in range(n):
    for j in range(0,i+1):
        print("# ",end="")
        
    
    print()
    
#print inverted inverted right angled triangle
print("Inverted Right angled pattern")
for i in range(n,0,-1):
    for j in range(1,i+1):
        print("# ",end="")
        
    
    print()
 

#print start pattern

print("Start pattern")
for i in range(n):
    for j in range(0,i+1):
        print("# ",end="")
        
    
    print()
    


for i in range(n,0,-1):
    for j in range(2,i+1):
        print("# ",end="")
        
    
    print()
 

```
